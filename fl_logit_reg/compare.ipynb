```python
import pandas as pd
```

## 对比

```python
with open('./results/res_center.txt', 'r') as f:
    res_center = []
    for line in f.readlines():
        line = line.strip().split(',')
        res_center.append(line)
res_center_df = pd.DataFrame(pd.array(res_center,dtype=float), columns=['epoch', 'loss_cen', 'acc_cen'])
```

```python
res_center_df.set_index('epoch', inplace=True)
```

```python
with open('./results/res_fl.txt', 'r') as f:
    res_fl = []
    for line in f.readlines():
        line = line.strip().split(',')
        res_fl.append(line)
res_fl_df = pd.DataFrame(pd.array(res_fl,dtype=float), columns=['epoch', 'loss_fl', 'acc_fl'])
```

```python
res_fl_df.set_index('epoch', inplace=True)
```

```python
res = pd.concat([res_center_df, res_fl_df], axis=1)
```
